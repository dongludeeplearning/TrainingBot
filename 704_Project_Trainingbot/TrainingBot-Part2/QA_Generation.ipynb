```python
import re
import textract
#read the content of pdf as text
text = textract.process('MOJOManual-English.pdf').decode('utf-8')
#use four space as paragraph delimiter to convert the text into list of paragraphs.
paras = (re.split('\s{4,}',text))
answer = ''
for payload in paras:
  x = {
            "input_text": payload
        }
  output = qg.predict_shortq(x)
  if 'questions' in output:
    #print(output['questions'])
    for ques in output['questions']:
      answer += ('text:'+ques['Question']+'  '+'labels:'+ques['Answer']+'  '+'episode_done:True\n')
print(answer)
```

```python
answer
```

```python
with open('qapairs.txt', 'w') as f:
    f.write(answer)
```
